importing the train data and the test data and convert them from csv to data frame

```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

# preprocessing the data
select the deatures that we need in training the model and the output
the null values should be susbtituted in the data with the mean values to avoid problems in training the model
replacing male and female types with numrical values

```python
Features_V1 = ["Pclass", "Sex", "Age", "Fare","Survived"]
train = train[Features_V1]

mean_value = train['Age'].mean()
train['Age'] = train['Age'].fillna(mean_value)
mean_value = train['Fare'].mean()
train['Fare'] = train['Fare'].fillna(mean_value)

train["Sex"].replace('female', 0, inplace=True)
train["Sex"].replace('male', 1, inplace=True)
```

# StratifiedShuffleSplit
in this section of the code we split the data into 80% training data and 20% validation data 
and keeping teh percentage of female and male data  in the training and validation data as same as teh original data

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(train, train["Sex"]):
    strat_train_set = train.loc[train_index]
    strat_test_set = train.loc[test_index]
```

# Features
spliting the features to input values and ouyput values for both training and validation

```python
Features = ["Pclass", "Sex", "Age", "Fare"]
X_train = strat_train_set[Features]
Y_train = strat_train_set["Survived"]
X_valdiation = strat_test_set[Features]
Y_valdiation = strat_test_set["Survived"]
```

# logistic regression
fiting the the data in the model

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, Y_train)
```

# score
model evaluation by calculating the accuracy of the model by getting the percentage the right predcitions 

```python
score = LR.score(X_valdiation, Y_valdiation)
print(score)
```

# preproccesing test data
as the training data we preprocessor the test data by removing the null values and cahnge female and male by numric values

```python
mean_value = test['Age'].mean()
test['Age'] = test['Age'].fillna(mean_value)
mean_value = test['Fare'].mean()
test['Fare'] = test['Fare'].fillna(mean_value)

test["Sex"].replace('female', 0, inplace=True)
test["Sex"].replace('male', 1, inplace=True)
```

# fitting all the data

```python
LR.fit(train[Features], train["Survived"])
```

# submission
change the data frames to csv files to be submitted 

```python
testV = test[Features]
test_preds = LR.predict(testV)

# The lines below shows how to save predictions in format used for competition scoring
# Just uncomment them.

output = pd.DataFrame({'PassengerId': test.PassengerId,
                       'Survived': test_preds})
print(output)
output.to_csv('submission.csv', index=False)
```
